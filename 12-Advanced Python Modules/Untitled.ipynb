```python
#Wap to read data from Student.csv file.
import csv 

with open("students.csv",'r') as f:
    d= csv.reader(f)
    for rows in d:
        print(rows)
```

```python
#wap to search the record from Student.csv according to the id.
with open("students.csv",'r') as f:
    d= csv.reader(f)
    next(f) #skip the header row
    k=0;
    input_id=int(input("Enter the id number if student"))
    for row in d:
        if int(row[0])==input_id:
            print(row)
            break
        else:
            print("Record not found")
```

```python
#wap to insert/add records in file 
fields = ['RollNo','first_name','gender']
f = open("students1.csv",'w')
d=csv.writer(f)
d.writerow(fields)
ch='y'
while ch=='y' or ch=='Y':
    rn = int(input("Enter roll number:"))
    first_Name = input("Enter first name")
    gender = input("Enter your gender")
    rec = [rn,first_Name,gender]
    d.writerow(rec)
    ch=input("Enter more records??(Y/N)")

f.close()
```
